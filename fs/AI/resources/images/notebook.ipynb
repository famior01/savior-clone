### **Change the label**

```python
# path = r'C:\\Product\\resources\\images'
# files = glob(path+'/'+'example'+'/labels/*')

def write_label(path_link, label_str):
    path  = path_link.replace('\\','/')
    label = label_str
    
    text = []
    
    # this will save modified text into text 
    try:
        txt = open(path, 'r')
    except FileNotFoundError:
        pass
    else:
        a= txt.readlines()
        total_lines = len(a)
        for i in range(total_lines):
            a[i] = a[i].split()
            a[i][0] = label
            string = ' '.join(a[i])
            text.append(string)
        print(text)

# This will write new text 
        writ = open(path, 'w')
        for i in range(total_lines):
            writ.writelines(text[i])
            writ.write('\n')
        writ.close()
```
